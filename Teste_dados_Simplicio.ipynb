**Remover as pontuações**

```python
import string

def remover_pontuacao(texto):
    # criando uma tabela de tradução para remover pontuações
    tabela = str.maketrans('', '', string.punctuation)
    # removendo as pontuações do texto usando a tabela de tradução
    texto_sem_pontuacao = texto.translate(tabela)
    return texto_sem_pontuacao

# teste da função com uma string
exemplo = 'Esta é uma frase de exemplo para remoção de pontuação:?!;,.'
resultado = remover_pontuacao(exemplo)
print(resultado)
```

```python
def analisar_sentimento_sem_dicionario(frase):
    sia = SentimentIntensityAnalyzer()
    scores = sia.polarity_scores(frase)

    return scores

def mapear_emocao_sem_dicionario(scores_sentimento):
    threshold_positivo = 1.268 #número encontrado com algoritmo genético
    threshold_negativo = -1.29 #número encontrado com algoritmo genético

    if scores_sentimento['compound'] > threshold_positivo:
        if scores_sentimento['pos'] > scores_sentimento['neg']:
            return 'alegria'
        else:
            return 'surpresa'
    elif scores_sentimento['compound'] > abs(threshold_negativo):
        if scores_sentimento['neg'] > scores_sentimento['pos']:
            return 'tristeza'
        else:
            return 'raiva'
    elif scores_sentimento['pos'] > scores_sentimento['neg']:
        return 'desprezo'
    elif scores_sentimento['neg'] > scores_sentimento['pos']:
        return 'medo'
    else:
        return 'neutro'
```

```python
dicionario_emocoes = {
    'alegria': ['feliz', 'alegre', 'contente', 'felíc', 'santos', 'obrigado'],
    'tristeza': ['triste', 'desanimado', 'choro'],
    'raiva': ['raivoso', 'irritado', 'furioso'],
    'nojo': ['nojento', 'repugnante', 'asco'],
    'surpresa': ['surpreso', 'espantado', 'chocado'],
    'desprezo': ['desprezível', 'desdenhoso', 'desprezo'],
    'medo': ['assustado', 'amedrontado', 'temeroso', 'amputar', 'medo', 'agulha']
}

dicionario_emocoes = {
    emocao: (" ".join(palavras)) for emocao, palavras in dicionario_emocoes.items()
}

def analisar_sentimento_com_dicionário(frase):
    sia = SentimentIntensityAnalyzer()
    scores = sia.polarity_scores(frase)
    return {k: v for k, v in scores.items() if k != 'compound'}

def mapear_emocao_com_dicionario(scores_sentimento):
    threshold_positivo = 0.0
    threshold_negativo = 1.0

    for emocao, palavras in dicionario_emocoes.items():
        if any(palavra in scores_sentimento for palavra in palavras):
            return emocao

    if scores_sentimento['pos'] > threshold_positivo:
        if scores_sentimento['pos'] > scores_sentimento['neg']:
            return 'alegria'
        else:
            return 'surpresa'
    elif scores_sentimento['neg'] > abs(threshold_negativo):
        if scores_sentimento['neg'] > scores_sentimento['pos']:
            return 'tristeza'
        else:
            return 'raiva'
    elif scores_sentimento['pos'] > scores_sentimento['neg']:
        return 'desprezo'
    elif scores_sentimento['neg'] > scores_sentimento['pos']:
        return 'medo'
    else:
        return 'neutro'
```

Lógica: procuramos combinar as palavras do dicionário com o texto, se encontrar o algoritmo devolve automaticamente a emoção correspondente, caso não encontre nenhuma palavra o algoritmo procura encaixar a frase em um dos casos de if-elif-else

**Detalhe de atenção - como ele funciona atribuindo polaridades e tentando encaixa-las na lógica serão necessários vários testes para encontrar a lógica e os thresholds perfeitos**

```python
from deep_translator import GoogleTranslator
import concurrent.futures

def traduzir(pergunta):
    translated = GoogleTranslator(source='auto', target='pt').translate(pergunta)
    return translated

def traduzir_perguntas(perguntas):
    translated_coluna = []
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = executor.map(traduzir, perguntas)
        for translated in results:
            translated_coluna.append(translated)
    return translated_coluna
```

```python
import json
def limpar_coluna(column):
    def parse_json(value):
        try:
            parsed_value = json.loads(value)
            cleaned_value = str(parsed_value).replace(':', '').replace('[', '').replace(']', '').replace('{', '').replace('}', '')
            return cleaned_value
        except json.JSONDecodeError:
            return value
    return column.apply(parse_json)
```

as emoções estão em [i][números pares]

os números estão em [i][números impares][2:-1]

```python
def encontrar_maiores_numeros(comparacao_dos_modelos):
    maiores_numeros_1 = []

    for linha in comparacao_dos_modelos:
        max_numero_1 = None
        max_emocao_1 = None

        for i in range(1, len(linha), 2):
            numero_1 = float(str(linha[i])[2:-1])
            if max_numero_1 is None or numero_1 > max_numero_1:
                max_numero_1 = numero_1
                max_emocao_1 = linha[i - 1]

        maiores_numeros_1.append([max_emocao_1, max_numero_1])

    return maiores_numeros_1
```

```python
maiores_numeros_1 = encontrar_maiores_numeros(comparacao_dos_modelos_2)
```
